```python
import random
import time

# Define the cards using dictionaries
card_deck = [
    {"Name": "Diablo", "Health": 100, "Attack": 90, "Defense": 60},
    {"Name": "Medusa", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Jester", "Health": 120, "Attack": 60, "Defense": 90},
    {"Name": "Troll", "Health": 150, "Attack": 40, "Defense": 94},
    {"Name": "Specter", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Mist", "Health": 100, "Attack": 75, "Defense": 65},
    {"Name": "Savage", "Health": 100, "Attack": 90, "Defense": 50},
    {"Name": "Marauder", "Health": 100, "Attack": 85, "Defense": 50},
    {"Name": "Wimp", "Health": 110, "Attack": 40, "Defense": 85},
    {"Name": "Sorcerer", "Health": 100, "Attack": 70, "Defense": 55},
]

# Number of cards for each player
deck_size = 5

# Assign random decks to the player and the opponent
player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample(card_deck, deck_size)

def display_card(card):
    """Helper function to display card details."""
    print(f"Name: {card['Name']}, Health: {card['Health']}, Attack: {card['Attack']}, Defense: {card['Defense']}")

def battle(player_card, opponent_card):
    """Function to handle the battle between two cards."""
    print(f"\nPlayer uses {player_card['Name']} to attack!")
    
    if player_card['Attack'] >= opponent_card['Health']:
        print(f"{player_card['Name']} defeats {opponent_card['Name']}!")
        return True
    
    while player_card['Attack'] > 0 and opponent_card['Health'] > 0:
        if opponent_card['Defense'] > 0:
            damage_to_defense = min(player_card['Attack'], opponent_card['Defense'])
            opponent_card['Defense'] -= damage_to_defense
            print(f"{player_card['Name']} attacks {opponent_card['Name']}! {damage_to_defense} damage done to defense!")
        else:
            damage_to_health = player_card['Attack']
            opponent_card['Health'] -= damage_to_health
            print(f"{player_card['Name']} attacks {opponent_card['Name']}! {damage_to_health} damage done to health!")

        time.sleep(1)  # Pause for a moment for effect
        if opponent_card['Health'] <= 0:
            print(f"{opponent_card['Name']} has been defeated!")
            return True
        else:
            print(f"{opponent_card['Name']}'s new stats: Health={opponent_card['Health']}, Defense={opponent_card['Defense']}")
            return False

# Game loop
while player_deck and opponent_deck:
    print("\nPlayer's cards:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}. {card['Name']} (Health: {card['Health']}, Attack: {card['Attack']}, Defense: {card['Defense']})")

    # Opponent draws a card
    opponent_card = random.choice(opponent_deck)
    print("\nOpponent draws:")
    display_card(opponent_card)

    # Player chooses a card to fight with
    player_choice = int(input("\nChoose a card from your deck by number: ")) - 1
    player_card = player_deck[player_choice]

    # Battle phase
    if battle(player_card, opponent_card):
        # Opponent card is defeated
        opponent_deck.remove(opponent_card)
    else:
        # Player card is defeated
        player_deck.remove(player_card)

    # Check for victory or defeat
    if not player_deck:
        print("\nYou have no cards left! The opponent wins!")
        break
    elif not opponent_deck:
        print("\nThe opponent has no cards left! You win!")
        break

    # Pause for next round
    time.sleep(3)
    print("\nA new round begins...\n")
```
